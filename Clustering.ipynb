```python
from sklearn.datasets import load_wine

wine_data = load_wine()
```

```python
import pandas as pd
```

```python
wine = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)
```

```python
wine
```

```python
import matplotlib.pyplot as plt
```

```python
from pandas.plotting import scatter_matrix
```

```python
scatter_matrix(wine.iloc[:, [0, 5]])

plt.show()
```

```python
X = wine[['alcohol', 'total_phenols']]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
```

```python
scale.fit(X)
```

```python
scale.mean_
```

```python
scale.scale_
```

```python
X_scale = scale.transform(X)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(X_scale)
```

```python
y_pred = kmeans.predict(X_scale)
```

```python
print(y_pred)
```

```python
kmeans.cluster_centers_
```

```python
plt.scatter(X_scale[:, 0], X_scale[:, 1], c=y_pred)

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='*', s=250, c=[0,1,2])

plt.xlabel('alcohol')
plt.ylabel('total phenols')
plt.title('k-means (k=3)')
plt.show()
```
